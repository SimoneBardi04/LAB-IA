```python
import os
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_curve, auc,
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Configurazione stile
sns.set_theme(style="whitegrid")
os.makedirs('plots', exist_ok=True)

# 1. Caricamento Dati
dataset_url = 'https://www.kaggle.com/datasets/blastchar/telco-customer-churn'
download_path = kagglehub.dataset_download("blastchar/telco-customer-churn")
dataset_path = os.path.join(download_path, "WA_Fn-UseC_-Telco-Customer-Churn.csv")

df = pd.read_csv(dataset_path)
print(f"Dataset caricato: {df.shape[0]} righe, {df.shape[1]} colonne.")
```

```python
# 2. Pulizia Dati e Preprocessing
# Rimuoviamo customerID
df = df.drop(columns=['customerID'])

# Pulizia TotalCharges (conversione a numerico e gestione valori vuoti)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'].str.strip(), errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0.0)

# Mappatura Target
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Definizione Features e Target
X = df.drop(columns=['Churn'])
y = df['Churn']

num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
cat_cols = [col for col in X.columns if col not in num_cols]

# Definizione Preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), cat_cols)
    ]
)

# 3. Suddivisione Training e Test (70/30, stratificato)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Preprocessing completato.")
print(f"Train set: {X_train.shape[0]} campioni")
print(f"Test set:  {X_test.shape[0]} campioni")
```

```python
from sklearn.tree import DecisionTreeClassifier

# 4. Decision Tree: Training and Optimization
dt_config = {
    'model': DecisionTreeClassifier(random_state=42),
    'params': {
        'classifier__max_depth': [None, 5, 10, 20],
        'classifier__min_samples_split': [2, 5, 10],
        'classifier__criterion': ['gini', 'entropy']
    }
}

# Pipeline base
dt_base_pipe = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', dt_config['model'])])
dt_base_pipe.fit(X_train, y_train)

# Grid Search
dt_opt_pipe = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', dt_config['model'])])
dt_grid = GridSearchCV(dt_opt_pipe, dt_config['params'], cv=5, scoring='f1', n_jobs=-1)
dt_grid.fit(X_train, y_train)

print(f"Decision Tree optimization complete.")
print(f"Best F1 Score: {dt_grid.best_score_:.4f}")
```

```python
# 5. Random Forest: Training and Optimization
# Ripristinato per garantire la coerenza con la valutazione finale
rf_config = {
    'model': RandomForestClassifier(random_state=42),
    'params': {
        'classifier__n_estimators': [50, 100, 200],
        'classifier__max_depth': [None, 10, 20],
        'classifier__min_samples_split': [2, 5, 10]
    }
}

# Pipeline base
rf_base_pipe = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', rf_config['model'])])
rf_base_pipe.fit(X_train, y_train)

# Grid Search
rf_opt_pipe = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', rf_config['model'])])
rf_grid = GridSearchCV(rf_opt_pipe, rf_config['params'], cv=5, scoring='f1', n_jobs=-1)
rf_grid.fit(X_train, y_train)

print(f"Random Forest optimization complete.")
print(f"Best F1 Score: {rf_grid.best_score_:.4f}")
```

```python
# Matrici di Confusione (DT e RF)
models_to_plot = {
    'Decision Tree': dt_grid.best_estimator_,
    'Random Forest': rf_grid.best_estimator_
}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for i, (name, model) in enumerate(models_to_plot.items()):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i], cbar=False,
                xticklabels=['Rimasto', 'Abbandonato'],
                yticklabels=['Rimasto', 'Abbandonato'])
    axes[i].set_title(f'Matrice di Confusione:\n{name}')
    axes[i].set_xlabel('Predetto')
    axes[i].set_ylabel('Reale')

plt.tight_layout()
plt.show()
```

```python
# 7. Final Comparison and ROC Curves (Decision Tree vs Random Forest)
models_data = {
    'Decision Tree': (dt_base_pipe, dt_grid.best_estimator_),
    'Random Forest': (rf_base_pipe, rf_grid.best_estimator_)
}

plt.figure(figsize=(10, 8))
colors = {'Decision Tree': 'orange', 'Random Forest': 'green'}

print(f"{'Modello':<20} | {'Stato':<12} | {'Accuracy':<8} | {'F1-Score':<8} | {'ROC-AUC':<8}")
print('-' * 75)

for name, (base, opt) in models_data.items():
    for label, model, style in [('Default', base, '--'), ('Ottimizzato', opt, '-')]:
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

        acc = accuracy_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_prob)

        print(f"{name:<20} | {label:<12} | {acc:.4f}   | {f1:.4f}   | {roc_auc:.4f}")

        fpr, tpr, _ = roc_curve(y_test, y_prob)
        plt.plot(fpr, tpr, label=f'{name} ({label}, AUC={roc_auc:.3f})',
                 color=colors[name], linestyle=style, linewidth=2 if label=='Ottimizzato' else 1)

plt.plot([0, 1], [0, 1], 'k--', label='Random (AUC=0.500)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Final ROC Curve Comparison: DT vs RF')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
